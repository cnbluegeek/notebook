# CSS3 笔记

## 1. 元素选择器
### 简单选择器

Type 直接用元素标签名

Id 使用 # + 元素的id属性名，id属性是唯一的

Class 使用 . + 元素的class属性名

### 属性选择器
E[foo], E元素中，设置了foo属性的元素

E[foo='bar'], E元素中foo属性取值含有'bar'的元素

E[foo~='bar'], E元素中foo属性取值精确等于‘bar’的元素

E[foo^='bar'], 以‘bar’开始的

E[foo$='bar'], 以‘bar’结束的

E[foo*='bar'], 以‘bar’为子串的

E[foo|='en'],以‘en’开头的

### 上下文选择器
e1 e2, e1的所有后代元素中的e2元素

e1>e2, e1的子元素中的e2元素

e1~e2, e1的兄弟元素e2

e1+e2, e1的紧邻兄弟元素e2

*, 所有元素

### 伪类选择器
结构化的伪类选择器：
e:first-child, e元素是其父元素的第一个子元素

e:last-child, e元素是其父元素的最后一个子元素

e:nth-child(k), e元素是其父元素的第k个子元素， 

```css
k取 2n+1, 表示奇数
k取 2n, 表示偶数
```

UI的伪类选择器：

a:visited, 访问过的

a:link, 没有访问过的

e:hover, 鼠标悬停在e元素上

e:focus, 获得焦点的e元素

e:checked, 单选或复选框按钮被选中的e元素

e:active, 元素被激活

e:target, 通过文档内链接导航到的e元素

### 伪元素选择器
E::first-letter, 选中E元素中的首字符

E::first-line, 选中E元素中的第一行

E::before, 在元素内容的开头添加虚拟内容

```css
E::before{
  content: "--"; 在E元素内容的开头加上"--"
}
```

E::after, 在元素的结尾添加虚拟内容

### 级联和继承
级联算法：

(1) 重要性: 加 !importance 的高于不加的

(2) 规则来源: 浏览器预设  < 用户设置  <  开发者设置(无 !importance, 加!importance则相反)

(2) 特指度: ID>Class>Type，三个分量I-C-T

(3) 规则顺序: 后出现的规则优先级高

## 8. CSS布局

### 1. 浮动与清除
float, 浮动

```css
left; 向左浮动
right; 向右浮动
```

clear,

```css
none; 不清除，默认
left; 清除左浮动
right; 清除右浮动
both; 清除左右浮动
```

浮动会导致脱离父元素，解决方法：

(1) 设置父元素 overflow:hidden; 不建议使用

(2)让父元素也浮动；不建议使用

(3)在最后的元素后面再添加一个空的子元素，并设置该空元素的clear属性。

(4)利用伪元素选择器，在最后添加一个空元素

```css
#div1::after{
  content: ".";
  clear: both;
  display: block;
  visibility: hidden;
  height:0;
}
```

### 2. 定位
position,

```css
static; 默认
relative; 相对定位，定位元素
absolute; 绝对定位，该元素脱离父元素，找祖先中最近的定位元素作为参考点
fixed; 固定定位，页面滚动时，其位置不动，相对于浏览器视窗定位
```

定位元素所具有的特点：

(1) 将非定位元素转换为定位元素

(2) 成为后代元素作绝对定位时的参照物

(3) 将元素的默认层级进行改变，也可由z-index属性设置
