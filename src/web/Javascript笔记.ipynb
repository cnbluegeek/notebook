```javascript
someNode.insertBefore(newNode, null); // 插入成为最后一个子节点
someNode.inertBefore(newNode, someNode.firstChild);//插入后成为第一个子节点
someNode.insertBefore(newNode, someNode.lastNode);//插入后成为倒数第二个子节点
```

删除：

```javascript
someNode.removeChild(someNode.firstNode);//移除第一个子节点
someNode.removeChild(someNode.lastNode);//移除最后一个子节点
```

克隆:

```javascript
var deepList = myList.cloneNode(true);//深拷贝
var shallowList = myList.cloneNode(false);//浅拷贝
```

效率：

innerHTML:每次都会改变document，因此最好先进行字符串拼接，然后在更新到节点中

```javascript
	//代码1
	for(var x=0; x<10; x++){
		var li = document.createElement("li");
		li.innerHTML = "List item " + x;
		listNode.appendNode(li);
	}

	//代码2，比代码1有效率
	var html = "";
	for(var x=0; x<10; x++){
		html += "<li>List item " + x + </li>;
	}
	listNode.innerHTML = html;
```

outerHTML

DocumentFragment

```javascript
	//代码3，比代码1和代码2的好处是，不会改变原先绑定在元素上的事件
	var frag = document.createDocumentFragment();
	for(var x=0; x<10; x++){
		var li = document.createElement("li");
		li.innerHTML = "List item " + x;
		frag.appendChild(li);
	}
	listNode.appendChild(frag);
```
