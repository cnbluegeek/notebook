# python包的安装

Python有大量的第三方库，通常我们根据需要选择安装一些，而不是像Matlab那样全装上。个性化的开发环境通常都需要手动安装，下面我们就讨论一下几种常见的Python包的安装方法。
### 1. 源码安装
源码安装需要先下载该Python包的源码，然后运行setup.py文件。安装代码如下：
```
python setup.py install
```

### 2. 工具安装
目前有pip, easy_install 和 distribute 等等第三方工具来进行Python包的自动安装。pip是对easy_install的取代，distribute是对setuptools的取代。

#### 1. pip
首先需要安装pip工具软件。
(1). 源代码方式：
```shell
# (替换为最新的包)
$ wget http://pypi.python.org/packages/source/p/pip/pip-0.7.2.tar.gz 
$ tar xzf pip-0.7.2.tar.gz
$ cd pip-0.7.2
$ python setup.py install
```
(2). easy_install安装:
```shell
$ easy_install pip
```
(3). get\_pip.py 脚本安装：
``` shell
$ curl -0 https://raw.github.com/pypa/pip/master/contrib/get-pip.py
$ sudo python get-pip.py
```

安装完成以后，将pip的路径设置到环境变量中，就可以使用pip安装Python包。\\
1. 安装package
```shell
  $ pip install Markdown (pip install --upgrade Markdown)
```
2. 列出安装的packages
```shell
  $ pip freeze
```
3. 安装特定版本的package
  通过使用==, >=, <=, >, <来指定一个版本号。
```shell
  $ pip install 'Markdown<2.0'
  $ pip install 'Markdown>2.0,<2.0.3'
```
4. 升级包\\
  升级包到当前最新的版本，可以使用-U 或者 --upgrade
```shell
  $ pip install -U Markdown
```
5. 卸载包
```shell
  $ pip uninstall Markdown
```
6. 查询包
```shell
  pip search "Markdown"
```
  安装后的包在python27/Lib/site-packages 目录下。

#### 2. easy\_install
使用easy_install工具需要先安装setuptools，下载地址<http://pypi.python.org/pypi/setuptools>.也可以用另外一种方式安装
``` shell
wget -q http://peak.telecommunity.com/dist/ez_setup.py
python ez_setup.py
```
安装好以后需要检查环境变量是否已经设置。Window下的路径为：%Python_Root%/Scripts/, Linux下的路径为：/usr/local/bin。
如果想安装MySQL数据的python接口，可以运行如下指令：easy_install MySQL-python;如果要删除该工具安装的软件包，可以运行如下指令：easy_install -m MySQL-python

#### 3. distribute
安装distribute的代码如下：
``` shell
$ curl -0 http://python-distribute.org/distribute_setup.py
$ sudo python distribute_setup.py
```
